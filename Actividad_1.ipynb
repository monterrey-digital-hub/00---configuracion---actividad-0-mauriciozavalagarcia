```python
import pyspark
from pyspark import SparkContext
sc=SparkContext()
```

```python
nums=sc.parallelize([1,2,3,4,5,6,7])
nums
```

```python
nums_delta=sc.parallelize([-76, -25, -100, -71, -51, -58, 99, -60, 47, -99, -19, 40, 60, 23, 74, -88, -86, -44, -4, 3, -7, 26, -70, 51, 90])
nums_delta
```

```python
squared = nums_delta.map(lambda x: x-1 if x > 0 else x*x).collect()
for nums_delta in squared:
  print('%i ' % (nums_delta))
```
